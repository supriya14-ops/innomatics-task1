```python
"""## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different tokens due to case sensitivity. This increases vocabulary size unnecessarily and can reduce model performance. Converting text to lowercase ensures consistency and avoids duplication.

---

### 2. What happens if stopwords are not removed?
If stopwords are not removed:
- The dataset becomes noisy
- Model training becomes less efficient
- Important words may get less importance
- Increased dimensionality in vectorization

---

### 3. When removing stopwords can be harmful?
1. Sentiment Analysis:
   Example: "I am not happy" → removing "not" changes meaning

2. Question Answering Systems:
   Example: "What is your name?" → removing "what" breaks intent

---

### 4. Difference between Stemming and Lemmatization

| Feature | Stemming | Lemmatization |
|--------|----------|---------------|
| Method | Rule-based cutting | Dictionary-based |
| Output | May not be real word | Always valid word |
| Example | "running" → "run" | "running" → "run" |
| Example | "better" → "better" | "better" → "good" |

Stemming is faster but less accurate, while lemmatization is slower but more meaningful."""
```

```python
import re
import string

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (length ≤ 2 except "no", "not")
    filtered_tokens = [
        word for word in tokens 
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(filtered_tokens)

    return filtered_tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in test_sentences:
    tokens, sentence = preprocess_text(text)

    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", sentence)

    results.append((tokens, sentence))
```

```python
import numpy as np

print("\n--- Token Analytics ---")

for i, (tokens, sentence) in enumerate(results):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    print(f"\nSentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
```

```python
from collections import Counter

all_tokens = []

for tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Run pipeline
output = full_pipeline(test_sentences)

print("\nPipeline Output:")
print(output)
```

```python
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

print("\n--- Edge Case Testing ---")

for text in edge_cases:
    tokens, sentence = preprocess_text(text)
    print("\nInput:", text)
    print("Tokens:", tokens)
    print("Sentence:", sentence)
```
