```python
from langchain_core.prompts import PromptTemplate

# Topic
topic = "Machine Learning"

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Generate outputs
teaching_output = teaching_template.format(topic=topic)
interview_output = interview_template.format(topic=topic)
story_output = story_template.format(topic=topic)

# Print results
print("Teaching Prompt:")
print(teaching_output)

print("\nInterview Prompt:")
print(interview_output)

print("\nStorytelling Prompt:")
print(story_output)
```

```python
from langchain_core.prompts import ChatPromptTemplate

# Step 1: Define role-based behavior
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_message = "You are an interviewer who asks challenging questions."
    elif role == "motivator":
        system_message = "You are a motivational speaker who inspires learning."
    else:
        system_message = "You are a helpful assistant."

    # Step 2: Create ChatPromptTemplate
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Talk about {topic}")
    ])

    # Step 3: Format prompt
    return chat_template.format_messages(topic=topic)


# Test Example
messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

```python
# Validation function
def validate_inputs(audience, tone):

    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Check audience
    if audience not in valid_audiences:
        print("Invalid audience! Setting default = 'beginner'")
        audience = "beginner"

    # Check tone
    if tone not in valid_tones:
        print("Invalid tone! Setting default = 'formal'")
        tone = "formal"

    return audience, tone


# Test cases
a1, t1 = validate_inputs("beginner", "fun")
a2, t2 = validate_inputs("kids", "friendly")  # invalid case

print(a1, t1)
print(a2, t2)
```

```python
from langchain_core.prompts import PromptTemplate

# Reuse validation function
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        audience = "beginner"

    if tone not in valid_tones:
        tone = "formal"

    return audience, tone


# Prompt Generator Function
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Ask interview questions about {topic} for a {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    # Step 3: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    # Step 4: Generate prompt
    return prompt.format(topic=topic, audience=audience, tone=tone)


# Test Example
result = generate_prompt("Neural Networks", "beginner", "fun", "storytelling")

print(result)
```

```python
from langchain_core.prompts import PromptTemplate

# Step 1: Create ONE reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Step 2: Test with 5 different inputs
cases = [
    {"topic": "AI", "audience": "beginner", "tone": "fun"},
    {"topic": "Python", "audience": "intermediate", "tone": "formal"},
    {"topic": "Data Science", "audience": "expert", "tone": "casual"},
    {"topic": "Machine Learning", "audience": "beginner", "tone": "formal"},
    {"topic": "Deep Learning", "audience": "expert", "tone": "fun"},
]

# Step 3: Run all cases
for case in cases:
    result = prompt.format(**case)
    print(result)
```
