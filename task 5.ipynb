```python
# Problem 1: Social Media – Post Engagement Analyzer

def post_engagement_analyzer(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


post_engagement_analyzer([200, 300, 550])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System

def medicine_stock_alert(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")


medicine_stock_alert(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_adequacy_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {int(average)}")
    print(f"Rainfall Status: {status}")


rainfall_adequacy_checker([60, 80, 75], 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def duplicate_account_detection(usernames):
    status = "Yes" if len(usernames) != len(set(usernames)) else "No"
    print(f"Duplicate Accounts Found: {status}")


duplicate_account_detection(["user1", "user2", "user1"])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker

def appointment_eligibility_checker(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")


appointment_eligibility_checker(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def premium_crop_price_filter(prices):
    premium = [price for price in prices if price > 2000]
    print(f"Premium Crops: {premium}")


premium_crop_price_filter([1500, 2500, 3200])
```

```python
# Problem 7: System Monitoring – Application Health Checker

def application_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {errors}")
    print(f"System Status: {status}")


application_health_checker(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def transaction_limit_checker(amount):
    status = "Approved" if amount <= 50000 else "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


transaction_limit_checker(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_eligibility_system(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day:
            attended += 1

    percentage = (attended / total_classes) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")


attendance_eligibility_system([True, True, False, True, True])
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill_analyzer(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Electricity Bill: ₹{bill}")
    print(f"Usage Level: {usage}")


electricity_bill_analyzer(250)
```
